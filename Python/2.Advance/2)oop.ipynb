## Abstraction

```python
class CoffeMachine():

    def __init__(self): # Initializes the machine with water and coffe beans.
        self.water = 100
        self.beans = 50

    def __heating_water(self):      # These are private methods (note the  __name): meaning they can't be accessed from outside. They are part of the internal process.
        print("Heating water...")

    def __grinding_beans(self):     # private method
        print("Grinding beans...")

    def __pouring_coffe(self):          # private method
        print("Pouring coffe...")

    def make_coffe(self):                   # This is the public method that users will call. It uses the private methods behind the scenes.
        if self.water >= 20 and self.beans >= 10:
            self.__heating_water()
            self.__grinding_beans()
            self.__pouring_coffe()
            self.water -= 20
            self.beans -=10
            print("‚òï Coffee is ready!")
        else:
            print("Not enough resources to make coffe.")

# Use the class
machine = CoffeMachine()
machine.make_coffe()        # This is the only method exposed to the user. The complexity of how coffe is made is abstracted away.
```

```python
class CoffeeMachine:
    def __init__(self):
        self.water = 1000
        self.beans = 100

    def __heating_water(self):
        print("Heating water...")

    def __grinding_beans(self):
        print("Grinding beans...")

    def __brew_coffee(self, coffee_type):
        print(f"Brewing your {coffee_type}...")

    def make_coffee(self):
        coffee_type = input("Which coffee do you want? (espresso or americano): ").lower()

        if coffee_type == "espresso":
            required_water = 50
            required_beans = 10
        elif coffee_type == "americano":
            required_water = 100
            required_beans = 25
        else:
            print("‚ùå Sorry, we don't have that type.")
            return

        if self.water >= required_water and self.beans >= required_beans:
            self.__heating_water()
            self.__grinding_beans()
            self.__brew_coffee(coffee_type)

            self.water -= required_water
            self.beans -= required_beans

            print(f"‚úÖ Your {coffee_type} is ready!")
        else:
            print("‚ùå Not enough resources. Please refill.")

    def refill(self):
        self.water += 500
        self.beans += 50
        print("üîÑ Machine refilled!")

# Test use
machine = CoffeeMachine()
machine.make_coffee()
machine.refill()
```

```python
class EspressoMachine():
    """Simulates an espresso machine that uses water and coffee beans."""

    def __init__(self):
        """Initialize machine with full resources."""
        self.water = 1000  # ml
        self.beans = 100   # grams

    # Internal steps (used only inside the class)
    def __heating_water(self):
        print(" üî• Heating water...")

    def __grinding_beans(self):
        print(" üåÄ Grinding beans...")

    def __brew_coffee(self, coffee_type):
        print(f" ‚òï Brewing your {coffee_type.title()}...")

    def make_coffee(self):
        """Ask user what type of coffee they want, then prepare it."""
        coffee_type = input("\nWhich coffee do you want? (espresso or americano): ").lower()

        if coffee_type == "espresso":
            required_water = 50
            required_beans = 10
        elif coffee_type == "americano":
            required_water = 100
            required_beans = 25
        else:
            print("‚ùå Sorry, we don't have that type of coffee.")
            return

        if self.water >= required_water and self.beans >= required_beans:
            print()
            self.__heating_water()
            self.__grinding_beans()
            self.__brew_coffee(coffee_type)

            # Subtract used resources
            self.water -= required_water
            self.beans -= required_beans

            print(f"‚úÖ Your {coffee_type.title()} is ready!\n")
        else:
            print("‚ùå Not enough resources. Please refill.\n")

    def refill(self):
        """Refill machine if resources are below 20%."""
        if self.water <= 200 or self.beans <= 20:
            self.water += 500
            self.beans += 50
            print("üîÑ Machine refilled!\n")
        else:
            print("‚ÑπÔ∏è  Water and beans are enough. No need to refill.\n")

    def status(self):
        """Display current resource levels."""
        print(f"üíß Water: {self.water}ml | üå± Beans: {self.beans}g")


# --- Use the machine ---
machine = EspressoMachine()

# 1. Check status
print("üîé Initial Machine Status:")
machine.status()

# 2. Make coffee
machine.make_coffee()

# 3. Try refill
machine.refill()

# 4. Final status
print("üîé Final Machine Status:")
machine.status()
```

## Encapsulation

```python
class BankAccount():
     
    def __init__(self):
        self.__balance = 0  # Creates a private variable to hold the balance. ‚õî It cannot be acessed directly outside the class -- this is a encapsulation in action.

    def deposit(self, amount):  # is a public method to safely add money.

        if amount > 0:  # check if amount is greater than 0 || if is valid, add it to __balance
            self.__balance += amount
            print(f"Deposited: ${amount}")

        else:
            print("Inavalid deposit amount.")   # Otherwise, shows an error message.
    
    def withdraw(self, amount):     # Public Method: used to subtract money. || Only works if you have enough money, if it is not enough show an error.

        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: ${amount}")
        
        else:
            print("Not enough funds.")

    def check_balance(self):        # A simple method that shows your balance.  || It gives controlled access to the private variable __balance.
        print(f"Current Balance: ${self.__balance}")

# Create an object of BankAccount:
account = BankAccount()

# Deposit:  (Calls the deposit() method to add $250.000)
account.deposit(250000)     

# Withdraw:     (Withdraws $50000 from the account (if balance is enough))
account.withdraw(50000)

# Check account:        (prints the current balance)
account.check_balance()

# Try to access directly to the account:
# Error: print(account.__balance)
```

```python
class BankAccount:
    # Constructor: runs when you create a new BankAccount
    def __init__(self):
        self.__balance = 0  # private attribute (encapsulated)

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount  # access and update private balance
            print(f"üí∞ Deposited: ${amount}")
        else:
            print("‚ùå Invalid deposit amount.")

    # Method to withdraw money
    def withdraw(self, amount):
        if amount <= self.__balance:  # only allow if enough funds
            self.__balance -= amount
            print(f"üí∏ Withdrawn: ${amount}")
        else:
            print("üö´ Not enough funds.")

    # Method to check balance
    def check_balance(self):
        print(f"üßæ Current Balance: ${self.__balance}")

# Create a new bank account
account = BankAccount()

# Deposit money
account.deposit(1000)

# Withdraw some money
account.withdraw(40)

# Check balance
account.check_balance()

# ‚ùå Trying to access private attribute directly:
# print(account.__balance)  # This will raise an AttributeError
```

```python
class Student():

    def __init__(self, name):
        self.name = name    # public attribute
        self.__grade = 0    # private attribute (encapsulated)

    def set_grade(self, value):
        """ Sets the grade if it is valid (0-100) """

        if 0 <= value <= 100:
            self.__grade = value
            print(f"Grade set to {value} for {self.name}.")

        else:
            print("Invalid grade. Must be between 0 and 100.")
    
    def get_grade(self):
        """Return the current grade."""
        print(f"{self.name}'s grade is {self.__grade}")

# create objetct:
studen_a = Student("Luis")
# Add grade
studen_a.set_grade(80)
# Show grade:
studen_a.get_grade()
```
